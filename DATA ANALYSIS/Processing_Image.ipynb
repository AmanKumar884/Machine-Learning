```python
# Resizing the image using pillow library
from PIL import Image
```

```python
img = Image.open('/content/gelding-bay-coat.jpg')
img_resize = img.resize((200,200))
```

```python
img_resize.save('horse_image_resize.jpeg')
```

```python
# Displaying the image from numpy array
img_res = mpimg.imread('/content/horse_image_resize.jpeg')
img_res_plot = plt.imshow(img_res)
plt.show()
```

```python
# Convert the RGB images to gray scale images using OpenCV(cv2)
import cv2
img = cv2.imread('/content/gelding-bay-coat.jpg')
```

```python
type(img)
```

```python
img.shape
```

```python
grayscale_image = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
```

```python
type(grayscale_image)
```

```python
grayscale_image.shape
```

```python
# from google.colab.patches import cv2_imshow
from google.colab.patches import cv2_imshow
```
